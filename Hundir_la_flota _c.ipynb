```python
import numpy as np
import random

# Símbolos para las celdas
celda_vacia = "-"        # Celda vacía, representa un espacio libre en el tablero
celda_barco = "O"        #Celda ocupada por un barco
celda_disparo = "X"      # Celda donde ha caído un disparo
celda_agua = "~"         # Celda donde ha caído un disparo yno ha tocado un barco

# Función para crear el tablero de juego
def crear_tablero():
    # Crear un tablero de 11x11 para incluir las coordenadas
    tablero = np.full((11, 11), celda_vacia, dtype=str)

    # Configurar las cooordenadas de la primera fila y columna
    for i in range(1, 11):
        tablero[0, i] = str(i)      # Números en la primera fila (1 a 10)
        tablero[i, 0] = chr(64 + i) # Letras en la primera columna (A a J), chr(64 + i), en este caso es Tabla de Códigos de los caracteres ASCII y el 65 es A y llega hasta la J

    return tablero  # Devolver el tablero configurado

# Función para colocar un barco en el tablero si es posible
def coloca_barco_plus(tablero, barco):
    tablero_copia = tablero.copy()  # Crear una copia temporal del tablero para no modificar el original directamente
    num_max_filas = tablero.shape[0]  # Número máximo de filas (11)
    num_max_columnas = tablero.shape[1]  # Número máximo de columnas (11)

    # Iterar sobre cada pieza del barco para verificar su colocación
    for pieza in barco:
        fila = pieza[0]   # Coordenada de fila
        columna = pieza[1]  # Coordenada de columna

        # Verificar si la pieza está fuera de los límites del tablero
        if fila < 0 or fila >= num_max_filas or columna < 0 or columna >= num_max_columnas:
            print(f"No puedo poner la pieza {pieza} porque se sale del tablero")
            return False  # Si alguna pieza está fuera del tablero, devolver False

        # Verificar si ya hay un barco o disparo en la posición
        if tablero_copia[pieza] == "O" or tablero_copia[pieza] == "X":
            print(f"No puedo poner la pieza {pieza} porque hay otro barco o disparo")
            return False  # Si hay un barco o disparo en la posición, devolver False

        tablero_copia[pieza] = "O"  # Colocar la pieza del barco en el tablero temporal

    return tablero_copia  # Si todas las piezas del barco se colocaron correctamente, devolver el tablero con el barco

# Función para crear un barco aleatorio y colocarlo en el tablero
def crea_barco_aleatorio(tablero, eslora=4, num_intentos=100):
    num_max_filas = tablero.shape[0]  # Número máximo de filas (11)
    num_max_columnas = tablero.shape[1]  # Número máximo de columnas (11)

    for _ in range(num_intentos):  # Intentos para colocar el barco
        barco = []  # Lista para almacenar las piezas del barco
        pieza_original = (random.randint(1, num_max_filas - 1), random.randint(1, num_max_columnas - 1))  # Elegir una pieza inicial aleatoria
        barco.append(pieza_original)  # Añadir la pieza original al barco

        # Elegir una orientación aleatoria para el barco: N, S, O, E (Norte, Sur, Oeste, Este)
        orientacion = random.choice(["N", "S", "O", "E"])
        fila, columna = pieza_original  # Asignar la fila y columna inicial

        # Generar las coordenadas del barco según la orientación
        for i in range(eslora - 1):  # Recorrer la longitud del barco, restando 1 porque ya hemos colocado la primera pieza
            if orientacion == "N":  # Si la orientación es Norte, disminuimos la fila
                fila -= 1
            elif orientacion == "S":  # Si la orientación es Sur, incrementamos la fila
                fila += 1
            elif orientacion == "E":  # Si la orientación es Este, incrementamos la columna
                columna += 1
            elif orientacion == "O":  # Si la orientación es Oeste, disminuimos la columna
                columna -= 1

            # Verificar si la pieza está fuera del tablero
            if fila < 1 or fila >= num_max_filas or columna < 1 or columna >= num_max_columnas:
                break  # Si la pieza está fuera de los límites, salir del bucle

            pieza = (fila, columna)  # Añadir la nueva pieza al barco
            barco.append(pieza)  # Añadir la pieza al barco

        # Intentar colocar el barco en el tablero
        tablero_copia = coloca_barco_plus(tablero, barco)
        if type(tablero_copia) == np.ndarray:  # Si el barco se coloca correctamente (la función devuelve un tablero)
            return tablero_copia  # Devolver el tablero con el barco colocado

    print("No pude colocar el barco después de varios intentos.")  # Si no se pudo colocar el barco después de varios intentos
    return tablero  # Devolver el tablero original si no se pudo colocar el barco

# Crear los tableros para el jugador y la máquina
tableros = {
    0: crear_tablero(),  # ID 0 es para el jugador
    1: crear_tablero()   # ID 1 es para la máquina
}

# Crear un barco aleatorio para el jugador (ID 0)
for eslora in [1, 1, 1, 1, 2, 2, 2, 3, 3, 4]:  # La lista con los tamaños de los barcos a colocar
    tableros[0] = crea_barco_aleatorio(tableros[0], eslora=eslora)

# Crear un barco aleatorio para la máquina (ID 1)
for eslora in [1, 1, 1, 1, 2, 2, 2, 3, 3, 4]:  # La lista con los tamaños de los barcos a colocar
    tableros[1] = crea_barco_aleatorio(tableros[1], eslora=eslora)
    
# Mostrar los tableros de los jugadores
print("Tablero del Jugador (ID 0):")
print(tableros[0])  # Imprimir el tablero del jugador

print("\nTablero de la Máquina (ID 1):")
print(tableros[1])  # Imprimir el tablero de la máquina
```
